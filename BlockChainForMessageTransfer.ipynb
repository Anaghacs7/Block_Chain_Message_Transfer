```python
from hashlib import sha256
import json

sender = 'Anu'

genesis_block = {'previous_hash': 'XYZ',
                 'index': 0,
                 'transactions': [], 'proof': 0}
blockchain = [genesis_block]
open_transactions = []
```

```python
def get_message():
    recipient = input('Enter your recipient: ')
    msg = input('Enter your message: ')
    return (recipient,msg)
```

```python
def proof_of_work():
    previous_hash = ''
    proof = 0
    last_block = blockchain[-1]
    for keys in last_block:
        previous_hash = previous_hash + str(last_block[keys])

    hashed = previous_hash + str(proof)
    hash = sha256(hashed.encode('utf-8')).hexdigest()
    while hash[0:2] != '00':
            guess_hash = previous_hash + str(proof)
            hash = sha256(guess_hash.encode('utf-8')).hexdigest()
            proof = proof + 1
            print(hash)
    return proof
```

```python
def hash_block(last_block):
    previous_hash = ''
    for keys in last_block:
        previous_hash = previous_hash + str(last_block[keys])
    hash = sha256(json.dumps(previous_hash).encode('utf-8')).hexdigest()
    return hash
```

```python


def mine_block():
    if open_transactions:
        last_block = blockchain[-1]
        hashed_block = hash_block(last_block)
        proof = proof_of_work()
        
        block = {
            'previous_hash': hashed_block,
            'index': len(blockchain),
            'transactions': open_transactions,
            'proof': proof
        }
        blockchain.append(block)
        print()
        print(blockchain)
        print()
        print(hashed_block)
        print()
        print(proof)
        print()
        return True
    else:
        print('\nNo blocks to mine\n')
```

```python
def verify_chain():
    
    for (index, block) in enumerate(blockchain):
        if index == 0:
            continue
        if block['previous_hash'] != hash_block(blockchain[index - 1]):
            print('Previous hash of block in blockchain: ' + block['previous_hash'])
            print('Calculated hash of previous block in blockchain: ' + hash_block(blockchain[index - 1]))
            print('\nBlock at Index ' + str(index))
            print(block)
            print('\nPrevious Block at Index ' + str(index-1))
            print(blockchain[index-1])
            return False

    return True
```

```python
def alter_block(ind, mod_msg):
    
    mod_transaction = {'sender': sender,
                       'recipient': recipient,
                       'message': mod_msg}
    
    blockchain[int(ind)] = {'previous_hash':'ABC',
                            'index':int(ind),
                            'transactions':mod_transaction,'proof':0}
```

```python
while True:
    print('P to print blockchain')
    print('1 to send message')
    print('2 to mine block')
    print('3 to alter block')
    print('4 to show pending transactions')
    I = input('\nEnter your choice: ')
    if I == 'P':
        print()
        print(json.dumps(blockchain,indent=3,separators=(", ", " = ")))
        print()
        
    if I == '1':
        data = get_message()
        recipient, msg = data
        transaction = {'sender': sender,
                       'recipient': recipient,
                       'message': msg}
        
        open_transactions.append(transaction)
        print()
        print(open_transactions)
        print()

    if I == '2':
        mine_block()
        open_transactions = []

    if I == '3':
        ind = input('Enter Block Index: ')
        mod_msg = input('Enter altered message: ')
        alter_block(ind,mod_msg)

    if I == '4':
        if open_transactions:
            print()
            print(open_transactions)
            print()
        else:
            print('\nNo pending transactions\n')

if not verify_chain():
        print('Invalid block')
        break
```
